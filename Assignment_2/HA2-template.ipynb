```python
### Your code here ###
G = nx.read_edgelist('Florence_marriages.txt',create_using=nx.Graph(),nodetype= str )
nx.draw(G)
plt.show()
```

### 1	Degree centrality

1.1	Compute and print the degree centrality of the nodes

```python
nx.degree_centrality(G)
```

1.2	Plot the network with node sizes proportional to their centrality

```python
### Your code here ###
#C is the degree centraily 
C = nx.degree_centrality(G)
#Multiplied by 300 so that it's displayed in a better way
nx.draw(G,with_labels=True,node_size=[v * 300 for v in C.values()])
plt.show()
```
